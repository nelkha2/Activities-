```python
# Import Dependencies
import pandas as pd
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
csv_path = "Resources/ted_talks.csv"
ted_df = pd.read_csv(csv_path)

ted_df.head()
```

```python
# Figure out the minimum and maximum views for a TED Talk
max_views = ted_df["views"].max()
max_views
min_views = ted_df["views"].min()
min_views
```

```python
# Create bins in which to place values based upon TED Talk views
# NE:
#bins = [50000, 10050000, 20050000, 30050000, 40050000, 50050000]

# Create labels for these bins
# NE:
#group_names = ["A", "B", "C", "D", "E"]

#Instructor 
bins = [0, 199999, 399999, 599999, 799999, 999999,
        1999999, 2999999, 3999999, 4999999, 50000000]
group_labels = ["199k", "399k", "599k", "799k", "999k", "2mil",
                "3mil", "4mil", "5mil", "50mil"]
```

```python
# Slice the data and place it into bins
#NE:
#ted_df["Group_Views"] = pd.cut(ted_df["views"], bins, labels=group_names)

ted_df["Group_Views"] = pd.cut(ted_df["views"], bins, labels=group_labels)
```

```python
# Place the data series into a new column inside of the DataFrame
ted_df.head(5)
```

```python
# Create a GroupBy object based upon "View Group"
ted_groupviews = ted_df.groupby("Group_Views")
ted_groupviews["comments","duration"].count()
#ted_groupviews.mean()

# Find how many rows fall into each bin
#ted_groupviews.count()

# Get the average of each column within the GroupBy object
#ted_groupviews.mean()
```
