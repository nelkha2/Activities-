```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share
# inlcluded the outer parameter b/c not one-to-one data - he used len to calculate rows of each table
merge_table = pd.merge(bitcoin_df, dash_df, on="Date", how="outer", suffixes=("_Bitcoin", "_Dash"))
merge_table
```

```python
# Rename columns so that they are differentiated
merge_table.columns

merge_table = merge_table.rename(columns={"Open_x": "Bitcoin Open", "High_x": "Bitcoin High", "Low_x": "Bitcoin Low",
                                      "Close_x": "Bitcoin Close", "Volume_x": "Bitcoin Volume", "Market Cap_x": "Bitcoin Market Cap"})

merge_table = merge_table.rename(columns={"Open_y": "Dash Open", "High_y": "Dash High", "Low_y": "Dash Low",
                                      "Close_y": "Dash Close", "Volume_y": "Dash Volume", "Market Cap_y": "Dash Market Cap"})
merge_table.head()

```

```python
# alternatively you can set your suffixes when the merge occurs
merge_table = merge_table.merge(
    bitcoin_df, dash_df, on="Date", suffixes=("_Bitcoin", "_Dash"))
```

```python
# Collecting best open for Bitcoin and Dash
max_Bitcoin_open = merge_table["Open_Bitcoin"].max()
max_Bitcoin_open
max_Dash_open = merge_table["Open_Dash"].max()
max_Dash_open

# Collecting best close for Bitcoin and Dash


# Collecting the total volume for Bitcoin and Dash
Vol_Bitcoin = merge_table["Volume_Bitcoin"].sum()
Vol_Bitcoin
Vol_Dash = merge_table["Volume_Dash"].sum()
Vol_Dash
```

```python
# Creating a summary DataFrame using above values
Data = {"Bitcoin_Open_Max":[max_Bitcoin_open], 
        "Dash_Open_Max":[max_Dash_open],
        "Bitcoin_Volume":[Vol_Bitcoin],
        "Dash_Volume":[Vol_Dash]}
Data_df = pd.DataFrame(Data)
Data_df
```
